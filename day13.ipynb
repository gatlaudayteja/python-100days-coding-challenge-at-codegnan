```python
#Nested loops
#printing tables using nested loops

for i in range(1,4):
    for j in range(1,21):
        print(f"{i} * {j} ={i*j}")
    print("**********************************************")
```

```python
#print table using function

n=int(input("Enter the number: "))
def print_table(n):
    for i in range(1,21):
        print(f"{n} * {i} ={n*i}")
print_table(n)
```

```python
n = 7

for i in range(n):
    for j in range(n):
        if (
            i == 0 or                      # top row
            i == n - 1 or                  # bottom row
            (i == 6 // 2) or               # middle row
            (j == 0 and i <= 5 // 2) or    # left vertical (top half)
            (j == n - 1 and i >= 6 // 2)   # right vertical (bottom half)
        ):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
